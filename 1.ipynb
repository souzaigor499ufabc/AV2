```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do problema
h = 10.0  # altura inicial em metros
v0 = 30.0  # velocidade inicial em m/s
angulo = 45.0  # ângulo de lançamento em graus
k = 0.1  # coeficiente de resistência do ar
g = 9.8  # aceleração da gravidade em m/s^2

# Parâmetros adicionais
m = 1.0  # massa do projétil em kg
k = 0.1  # coeficiente linear de resistência do ar

# Função para calcular a aceleração devido à resistência do ar (modelo linear)
def resistencia_ar(v):
    return -k * v / m

# Converter o ângulo para radianos
angulo_rad = np.radians(angulo)

# Componentes horizontal e vertical da velocidade inicial
v0x = v0 * np.cos(angulo_rad)
v0y = v0 * np.sin(angulo_rad)

# Tempo de voo máximo
t_max = 2.0 * v0y / g

# Inicializar arrays para os resultados
x = np.zeros(100)
y = np.zeros(100)
v = np.zeros(100)
a = np.zeros(100)

# Calcular a trajetória com resistência do ar (modelo linear)
for i in range(100):
    t = t_max * i / 99.0
    x[i] = v0x * t
    y[i] = h + v0y * t - 0.5 * g * t**2
    v[i] = np.sqrt(v0x**2 + (v0y - g * t)**2)
    a_resistencia = resistencia_ar(v[i])  # Resistência do ar (modelo linear)
    a[i] = g + a_resistencia  # Aceleração total (gravidade + resistência do ar)

# Funções para plotar os gráficos
def plot_trajetoria(x, y):
    with open('trajetoria.txt', 'w') as f:
        for xi, yi in zip(x, y):
            f.write(f"{xi} {yi}\n")
    plt.plot(x, y)
    plt.title("Trajetória do projétil")
    plt.xlabel("Distância horizontal (m)")
    plt.ylabel("Altura (m)")
    plt.grid()
    plt.show()

def plot_velocidade(x, v):
    with open('velocidade.txt', 'w') as f:
        for xi, vi in zip(x, v):
            f.write(f"{xi} {vi}\n")
    plt.plot(x, v)
    plt.title("Velocidade do projétil")
    plt.xlabel("Distância horizontal (m)")
    plt.ylabel("Velocidade (m/s)")
    plt.grid()
    plt.show()

def plot_aceleracao(x, a):
    with open('aceleracao.txt', 'w') as f:
        for xi, ai in zip(x, a):
            f.write(f"{xi} {ai}\n")
    plt.plot(x, a)
    plt.title("Aceleração do projétil")
    plt.xlabel("Distância horizontal (m)")
    plt.ylabel("Aceleração (m/s²)")
    plt.grid()
    plt.show()

# Plotar os gráficos
plot_trajetoria(x, y)
plot_velocidade(x, v)
plot_aceleracao(x, a)
```

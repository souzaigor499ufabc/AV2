```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do problema
g = 9.8  # aceleração da gravidade em m/s²
altura_cesta = 3.05  # altura da cesta em metros
altura_inicial = 2.00  # altura inicial do lançamento em metros
distancia_cesta = 10.0  # distância horizontal até a cesta em metros
angulo = 40.0  # ângulo de lançamento em graus
v0 = 15.0  # velocidade inicial em m/s (ajuste conforme necessário)

# Converter o ângulo para radianos
angulo_rad = np.radians(angulo)

# Componentes horizontal e vertical da velocidade inicial
v0x = v0 * np.cos(angulo_rad)
v0y = v0 * np.sin(angulo_rad)

# Tempo total de voo (estimado para cálculo)
t_total = 2 * v0y / g
t = np.linspace(0, t_total, 500)  # Dividir o tempo em 500 pontos

# Funções horárias para posição
x = v0x * t
y = altura_inicial + v0y * t - 0.5 * g * t**2

# Componentes da velocidade em função do tempo
vx = np.full_like(t, v0x)  # Velocidade horizontal é constante
vy = v0y - g * t  # Velocidade vertical varia com o tempo

# Componentes da aceleração em função do tempo
ax = np.zeros_like(t)  # Aceleração horizontal é zero (sem resistência do ar)
ay = np.full_like(t, -g)  # Aceleração vertical é constante e igual a -g

# Encontrar a altura máxima
altura_maxima_idx = np.argmax(y)  # Índice da altura máxima
x_altura_maxima = x[altura_maxima_idx]  # Posição x correspondente
y_altura_maxima = y[altura_maxima_idx]  # Altura máxima

# Velocidade e aceleração na altura máxima
vx_altura_maxima = vx[altura_maxima_idx]
vy_altura_maxima = vy[altura_maxima_idx]
ax_altura_maxima = ax[altura_maxima_idx]
ay_altura_maxima = ay[altura_maxima_idx]

# Resultados
print("a) Posição x e y da bola em função do tempo:")
for i in range(len(t)):
    print(f"t = {t[i]:.2f} s, x = {x[i]:.2f} m, y = {y[i]:.2f} m")

print("\nb) Componentes x, y da velocidade e aceleração em função do tempo:")
for i in range(len(t)):
    print(f"t = {t[i]:.2f} s, vx = {vx[i]:.2f} m/s, vy = {vy[i]:.2f} m/s, ax = {ax[i]:.2f} m/s², ay = {ay[i]:.2f} m/s²")

print("\nc) Posição x onde a altura é máxima:")
print(f"x = {x_altura_maxima:.2f} m, y = {y_altura_maxima:.2f} m")

print("\nd) Componentes x, y da velocidade e aceleração na altura máxima:")
print(f"vx = {vx_altura_maxima:.2f} m/s, vy = {vy_altura_maxima:.2f} m/s")
print(f"ax = {ax_altura_maxima:.2f} m/s², ay = {ay_altura_maxima:.2f} m/s²")    
```

```python
plt.figure(figsize=(8, 6))
plt.plot(t, y, label="Altura (y) em função do tempo (t)", color="blue")
plt.axhline(y=altura_cesta, color="red", linestyle="--", label="Altura da cesta (3.05 m)")
plt.axvline(t[altura_maxima_idx], color="green", linestyle="--", label="Tempo na altura máxima")
plt.scatter(t[altura_maxima_idx], y_altura_maxima, color="black", label="Altura máxima", zorder=5)
plt.title("Altura da bola em função do tempo")
plt.xlabel("Tempo (s)")
plt.ylabel("Altura (m)")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_altura(h0, t, g):
    """Calcula altura em fun√ß√£o do tempo para queda livre"""
    altura = h0 - 0.5 * g * t**2
    return np.maximum(altura, 0)  # garante que altura n√£o seja negativa

def tempo_queda(h0, g):
    """Calcula tempo total de queda"""
    return np.sqrt(2 * h0 / g)

# Constantes de gravidade (m/s¬≤)
G_TERRA = 9.81
G_MARTE = 3.72
G_LUA = 1.62

# Entrada do usu√°rio
h0 = float(input("Digite a altura inicial (m): "))
t_usuario = float(input("Digite o tempo desejado (s): "))

# a) Altura para um tempo espec√≠fico
altura_t = calcular_altura(h0, t_usuario, G_TERRA)
print(f"\nAltura ap√≥s {t_usuario}s: {altura_t:.2f}m")

# b) Posi√ß√£o em fun√ß√£o do tempo at√© atingir o solo
t_total = tempo_queda(h0, G_TERRA)
t = np.arange(0, t_total + 1, 1)
y = calcular_altura(h0, t, G_TERRA)

plt.figure(figsize=(10, 6))
plt.plot(t, y, 'b-', label='Terra')
plt.title(f'Queda Livre - Altura Inicial: {h0}m')
plt.xlabel('Tempo (s)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.legend()
plt.ylim(bottom=0)  # define limite inferior do gr√°fico como 0
plt.show()

# c) Compara√ß√£o entre Terra, Marte e Lua
t_terra = np.arange(0, tempo_queda(h0, G_TERRA) + 1, 1)
t_marte = np.arange(0, tempo_queda(h0, G_MARTE) + 1, 1)
t_lua = np.arange(0, tempo_queda(h0, G_LUA) + 1, 1)

y_terra = calcular_altura(h0, t_terra, G_TERRA)
y_marte = calcular_altura(h0, t_marte, G_MARTE)
y_lua = calcular_altura(h0, t_lua, G_LUA)

plt.figure(figsize=(10, 6))
plt.plot(t_terra, y_terra, 'b-', label='Terra')
plt.plot(t_marte, y_marte, 'r-', label='Marte')
plt.plot(t_lua, y_lua, 'g-', label='Lua')
plt.title(f'Compara√ß√£o da Queda Livre - Altura Inicial: {h0}m')
plt.xlabel('Tempo (s)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.legend()
plt.ylim(bottom=0)
plt.show()
```
